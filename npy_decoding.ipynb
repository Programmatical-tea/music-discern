```python
import tensorflow as tf
```

```python
int_file = tf.io.read_file("int.npy")
float_file = tf.io.read_file("float.npy")
```

```python
int_file
```

```python
float_file
```

```python
def process(raw, option, sizee, shapee): #map_func
    #extract the 2d array and the label
    if option == 1:  img = tf.io.decode_raw(raw,tf.float32)
    else:
        img = tf.io.decode_raw(raw,tf.int32)
    print("----------------------------------------------------------------------------------------------------------- decoded")
    print(img)
    img = tf.slice(img,begin=[1024//32],size=sizee) # begin <- 1024//complex64
    print("----------------------------------------------------------------------------------------------------------- sliced")
    print(img)
    img = tf.reshape(img,shapee)
    print("----------------------------------------------------------------------------------------------------------- reshaped")
    print(img)
    return img
```

```python
process(int_file,0,[9],(3,3))
print(A)
```

```python
process(float_file,1,[9],(3,3))
print(B)
```

```python
def make_and_decode_float(count,array,size,shape):
    np.save("sep_"+str(count), array, allow_pickle=True)
    img = tf.io.read_file("sep_"+str(count)+".npy")
    a = process(img,1,size,shape)
    print("----------------------------------------------------------------------------------------------------------- the original array")
    print(array)
    print("----------------------------------------------------------------------------------------------------------- difference")
    print(a.numpy()-array)
```

```python
make_and_decode_float(count,B,[9],(3,3))
count = count + 1
```
